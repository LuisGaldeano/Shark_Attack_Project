Lo primero que voy a hacer es eliminar los espacios

```python
shark['age'] = shark['age'].str.replace(' ', '')
```

Como esta culumna va a ser numérica para elimnar los nulos los sustituyo por un proxy numérico que me permita modificarlos en el futuro. En este caso lo voy a sustituir po -1, ya que la edad nunca puede ser negativa

```python
shark.age.fillna(-1, inplace=True)
```

```python
grafico(shark)
```

La siguiente columna a revisar es la de time

```python
shark['time'].isna().mean() * 100 # % de datos que son nulos
```

```python
shark['time'][:20]
```

```python
shark['time'] = shark['time'].replace('h', ':', regex=True)
```

```python
shark['time'][:20]
```

```python
shark['time1'] = shark['time'].str.extract(r'(\d{2}:\d{2})', expand=False)
```

```python
times = {
    'Afternoon': '18:00',
    'afternoon': '18:00',
    'Morning': '09:00',
    'morning': '09:00',
    'Night': '00:00',
    'night': '00:00',
    'Evening': '21:00',
    'evening': '21:00',
}

for key, value in times.items():
    shark['time'] = shark['time'].str.replace(key, value, case=False)

shark['time'][:20]
```

```python
shark['time1'][:20]
```

Llegados a este punto, como no puedo obtener más horas, elimino los valores nulos con un proxy 25:00, elimino la columna time y dejo solo la columna time1 que la renombro como time

```python
shark.time1.fillna('25:00', inplace=True)
```

```python
shark.drop('time', axis=1, inplace=True)
shark.rename(columns={'time1': 'time'}, inplace=True)
```

```python
def filas_con_nulos(df, columna):
    valores_nulos = df[columna].isna()

    filas = df[valores_nulos]

    return filas
```

```python
import pandas as pd
import numpy as np
import re
import zipfile
import pylab as plt    
import seaborn as sns
import sys
from datetime import datetime, time
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style()

from functions import *

pd.set_option('display.max_columns', None)  # me muestre todas las columnas
```

```python
shark=pd.read_csv('../data/sharkattack_análisis.csv')
```

```python
shark.head()
```

Primero voy a analizar las actividades que se ralizaban durante el ataque para ver que actividades son más susceptibles de sufrir ataques de tiburones

```python
filtro = shark[shark['activity'] != 'other']
```

```python
filtro['activity'].value_counts().plot(kind='bar')
```

Como se puede observar, el surf y la pesca son las dos actividades con un mayor riesgo de sufrir un ataque de tiburón

---

Lo siguiente que voy a analizar es el género más susceptible de sufrir un ataque

```python
filtro = shark[shark['sex'] != 'u']
```

```python
filtro['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

También se observa que los hombres sufren muchos más ataques de tiburones que las mujeres

---

Lo siguiente que me gustaría analizar es si hay alguna época del año que sea más propicia a sufrir ataques de tiburón

```python
filtro = shark[shark['mes'] != 0]
```

```python
filtro['mes'].value_counts().plot(kind='bar')
```

Se puede observar, que los meses más propensos a sufrir un ataque de tiburón son los meses de verano Julio, Agosto y Septiembre

---

Lo último que me gustaría analizar por separado es el momento del día en el que es más probable sufrir un ataque de tiburón

```python
filtro = shark[shark['hora'] != 'unknown']
```

```python
filtro['hora'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

Como se puede observar, la mayoría de ataques, con más del 60% se producen por la tarde

```python
filtro = shark[(shark['species'] != 'unknown shark') & (shark['species'] != 'other shark')]
```

```python
filtro['species'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

Se puede observar que el tiburón blanco es el que más ataques realiza con más de la mitad de ellos.
  
Le sigue el tiburón tigre con el 22% y el tiburón toro con el 14%

# Relaciones

```python
filtro = shark[(shark['hora'] != 'unknown') & (shark['activity'] != 'other')]
hora_actividad = filtro.groupby(['hora', 'activity']).size().unstack()
```

```python
hora_actividad.plot(kind='bar', stacked=True)
```

```python
filtro = shark[(shark['mes'] != 0) & (shark['activity'] != 'other')]
relacion = filtro.groupby(['mes', 'activity']).size().unstack()
relacion.plot(kind='bar', stacked=True)
```

```python
filtro = shark[(shark['species'] != 'unknown shark') & (shark['species'] != 'other shark') & (shark['mes'] != 0)]
relacion = filtro.groupby(['species', 'mes']).size().unstack()
relacion.plot(kind='bar', stacked=True, figsize=(10, 6))
```

```python
filtro = shark[(shark['species'] != 'unknown shark') & (shark['species'] != 'other shark') & (shark['activity'] != 'other')]
relacion = filtro.groupby(['species', 'activity']).size().unstack()
relacion.plot(kind='bar', stacked=True, figsize=(10, 6))
```

# Conclusiones

Tras analizar todos los gráficos, se puede llegar a las siguientes conclusiones en relaición al análiss propuesto:

1. Los tiburones blancos son los que más ataques realizan especialmente a surfistas y a pescadores
2. Por otro lado, no hay una relación directa entre los ataques de tiburones y la época del año, ya que los ataques se distribuyen de manera uniforme. Sin embargo si que los ataques son ligeramente más predominantes en los meses veraniegos
3. Sin embargo, si que hay una correlación entre el momento del día y los ataques, ya que la mayoría de los ataques se producen por la tarde seguido de las mañanas. Horarios donde se suele practicar el surf con mayor intensidad
  

# Como conclusión se puede afirmar que si practicas surf, especialmente entre las 12:00 y las 19:00 tienes más probabilidades de que te ataque un tiburón, que probablemente sea blanco
